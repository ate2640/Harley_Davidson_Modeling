```python
import pandas as pd
df = pd.read_csv('Harley_Dataset.csv')
```

```python
df.head()
```

```python
df.drop(['Order Number', 'Store Name', 'Order Date'], axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, columns=['Product Name', 'Payment Mode', 'City', 'Country'], drop_first=False, dtype=int)
```

```python
Order_Profit = df["Quantity"] * df["Price"]
df["Order_Profit"] = Order_Profit
```

```python
df.head()
```

```python
df.isna().any().sum()
```

```python
df.info()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
 
 
# Initialise the Scaler
scaler = StandardScaler()
 
# To scale data
scaler.fit(df)
```

```python
>>> import pandas as pd
>>> from sklearn.model_selection import train_test_split
>>> from sklearn.datasets import load_iris

y = df.Order_Profit
x = df.drop('Order_Profit', axis = 1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train.head()
```

# Model Testing

## Linear Regression

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot

LRmodel = LinearRegression()
```

```python
LRmodel.fit(x_train,y_train)
```

```python
LRmodel.score(x_test,y_test)
```

```python
print("The Linear Regression model predicts the profit column with", (round(LRmodel.score(x_test,y_test),4)*100),"% success")
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
RFC_model = RandomForestClassifier(random_state=42, class_weight='balanced', n_estimators=1000).fit(x_train,y_train)
RFC_model.score(x_test,y_test)
```

```python
print("The Random Forest Classifier model predicts the profit column with", (round(RFC_model.score(x_test,y_test),4)*100),"% success")
```
